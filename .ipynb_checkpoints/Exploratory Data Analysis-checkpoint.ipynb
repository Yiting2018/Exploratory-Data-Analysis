```python
start = datetime.datetime(2016, 1, 1)
end = datetime.datetime(2017, 1, 1)
```

```python
# Bank of America
BAC = data.DataReader("BAC", 'iex', start, end)

# CitiGroup
C = data.DataReader("C", 'iex', start, end)

# Goldman Sachs
GS = data.DataReader("GS", 'iex', start, end)

# JPMorgan Chase
JPM = data.DataReader("JPM", 'iex', start, end)

# Morgan Stanley
MS = data.DataReader("MS", 'iex', start, end)

# Wells Fargo
WFC = data.DataReader("WFC", 'iex', start, end)
```

```python
BAC.reset_index(inplace=True)
C.reset_index(inplace=True)
GS.reset_index(inplace=True)
JPM.reset_index(inplace=True)
MS.reset_index(inplace=True)
WFC.reset_index(inplace=True)
```

```python
# Could also do this for a Panel Object
df = data.DataReader(['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC'],'iex', start, end)
```

##### Draw the closing prices of BAC

```python
import matplotlib.pyplot as plt

# your code here
# Draw Plot
plt.figure(figsize=(100,30))
plt.plot(pd.to_datetime(BAC.date), BAC.close, color='tab:red')
# Decoration
plt.xticks(fontsize=32)
plt.yticks(fontsize=32)
plt.title("Closing Prices for Bank of America", fontsize=42)
plt.grid(axis='both')

# Remove borders|
plt.gca().spines["top"].set_alpha(0.0)    
plt.gca().spines["bottom"].set_alpha(0.3)
plt.gca().spines["right"].set_alpha(0.0)    
plt.gca().spines["left"].set_alpha(0.3)
plt.show()
```

##### Append all the data sets

```python
# your code here
banks = pd.concat([BAC, C, GS, JPM, MS, WFC])
```

##### Derive the average closing price for each bank

```python
# your code here
banks_grouped = banks.groupby("Company")
close_by_company = banks_grouped.close.mean()
close_by_company
```


** What is the max Close price for each bank's stock throughout the time period?**

```python
# your code here
max_close_by_company = banks_grouped.close.max()
max_close_by_company
```

** Create a new dataframe called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

** We can use pandas pct_change() method on the Close column to create a new dataframe representing this return value. Use .groupby().**

```python
# your code here
banks_list = [BAC, C, GS, JPM, MS, WFC]
for b in banks_list:
    b["returns"] = b.close.pct_change(periods=1)
banks_new = pd.concat([BAC, C, GS, JPM, MS, WFC])
```

** Using this returns DataFrame, figure out on what dates each bank stock had the best and worst single day returns. Did anything significant happen that day?**

```python
# your code here
banks_grouped = banks_new.groupby("Company")
min_returns_by_company = banks_grouped.returns.min()
max_returns_by_company = banks_grouped.returns.max()
```

```python
min_returns_by_company
```

```python
max_returns_by_company
```

```python
for name in min_returns_by_company.index:
    returns_value = min_returns_by_company[name]
    date = list(banks_new[(banks_new.Company == name) & (banks_new.returns == returns_value)].date)[0]
    print("Min returns for {} happened on {} with return of {}".format(name, date, np.round(returns_value, 3)))
```

```python
for name in max_returns_by_company.index:
    returns_value = max_returns_by_company[name]
    date = list(banks_new[(banks_new.Company == name) & (banks_new.returns == returns_value)].date)[0]
    print("Max returns for {} happened on {} with return of {}".format(name, date, np.round(returns_value, 3)))
```

** Please state here what you have noticed. Did anything significant happen in that time frame? **

It seems like the minimum returns all happened on June 24th, 2016.

That is the date British Prime Minister David Cameron resigns after the **UK votes to leave the EU!**

The maximum returns happened on Feb. 12th, 2016 (BOA, Citi, JPMorgan) and Nov. 9th/10th, 2016 (GoldmanSachs, MorganStanley, WellsFargo).
Based on our research, during the time of 2016-02-12, the stock market should be sinking instead of blooming. Our theory is the economic was simply recovering. On the date of 2016-11-09, which is the United States presidential election and general elections day, the banks generated the highest return.

### Moving Averages

** Please derive the moving averages for these stocks in the year 2016. Use .rolling() in pandas to get the rolling average calculation. ** 



```python
# your code here
# your code here
banks_list = [BAC, C, GS, JPM, MS, WFC]
for b in banks_list:
    b["moving_avg"] = b.close.rolling(window=30).mean()
banks_new = pd.concat([BAC, C, GS, JPM, MS, WFC])
```

** Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2016**

```python
# your code here
plt.figure(figsize=(20,15), dpi= 80)
sns.lineplot(data=banks_new[banks_new.Company == "BOA"].moving_avg, color="red", label="BOA", alpha=.4)
# Decoration
plt.title('Moving Average by Company', fontsize=22)
plt.legend()
plt.show()
```
